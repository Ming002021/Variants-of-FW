```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_Iter = pd.read_excel (r'/Users/Mindy/Desktop/Debug/ForGraph.xlsx',sheet_name='Iter') 
Iter=pd.DataFrame(data_Iter)

Iter

 
```

```python
BoxQP_Iter =Iter[Iter["Instance_Type"]=="RandQP&Globallib&CUTEr"]
BoxQP_e4_Iter=BoxQP_Iter [BoxQP_Iter ["Error_Type"]=="e4"]
BoxQP_e4_Iter
```

```python
sns.set(style="darkgrid")
plt.figure(figsize = (11,8))
sns.swarmplot(x='Variant', y='Number Iteration',hue="Start Point",data = BoxQP_e4_Iter,dodge=True)
#sns.boxplot(x='Variant', y='Number Iteration',hue="Start Point",data = BoxQP_e4_Iter)
plt.xlabel('Variants of Frank-Wolfe Algorithm')
plt.ylabel('Running Time')
plt.title(f' The Running Time in seconds of GeneralQP instances and 'r'$\epsilon=0.0001$',fontsize = 18)
plt.legend(title='Starting Point Method', loc='upper left')
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize = (11,8))
sns.boxplot(x='Variant', y='Number Iteration',hue="Start Point",data = BoxQP_e4_Iter)
plt.xlabel('Variants of Frank-Wolfe Algorithm')
plt.ylabel('Number of Iteration')
plt.title(f' The optimality gap% of GeneralQP instances and 'r'$\epsilon=0.0001$',fontsize = 18)
plt.legend(title='Starting Point Method', loc='upper left')
plt.show()
```
